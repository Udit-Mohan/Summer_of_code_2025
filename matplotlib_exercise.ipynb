Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
x = [1,2,3,4]
y = [1,2,3,4]
ax1.plot(x,y,color = 'blue')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(x,y,color = 'red')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

# Exercise 3


Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]



```python
fig = plt.figure()
ax1 = fig.add_axes( [0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
x = list(range(1,100))
y = [i**2 for i in x]
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.plot(x,y,color = 'blue')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim(20,22)
ax2.set_ylim(400,500)
ax2.set_title('Zoom')
ax2.plot(x,y,color = 'blue')
```

```python
fig,axes = plt.subplots(nrows = 1 , ncols = 2)
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Z')
x = list(range(1,1001))
y = [i**2 for i in x]
x2 = list(range(101))
y2 = [i*2  for i in x2]
axes[1].plot(x,y,color = 'red',linewidth = 4)
axes[0].plot(x2,y2,linestyle = 'dotted',linewidth = 4,color = 'blue')
plt.tight_layout()
plt.show()
```

```python

fig,axes = plt.subplots(nrows = 1 , ncols = 2,figsize = (10,3))
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Z')
x = list(range(1,1001))
y = [i**2 for i in x]
x2 = list(range(101))
y2 = [i*2  for i in x2]
axes[1].plot(x,y,color = 'red',linewidth = 4)
axes[0].plot(x2,y2,linestyle = 'dotted',linewidth = 4,color = 'blue')
plt.tight_layout()
plt.show()
```
